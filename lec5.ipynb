```python
import pandas as pd
import plotly.graph_objects as go


data = pd.read_csv('https://raw.githubusercontent.com/MoloZzz/nn/main/dataSetSalary.csv')


data.head(10)
```

```python
# Побудова діаграми розсіювання
fig = go.Figure(data=go.Scatter(x=data['lowest_salary'], y=data['highest_salary'],
                                mode='markers',
                                marker=dict(size=10, opacity=0.5),
                                text=data['country_name']))

# Налаштування вигляду діаграми
fig.update_layout(title='Співвідношення найнижчої та найвижчої зарплати по країнах',
                  xaxis_title='нижня межа зарплатні',
                  yaxis_title='верхня межа зарплатні')

# Ініціалізація режиму відображення у зошиті
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.io as pio

pio.renderers.default = "colab"

# Відобразіть діаграму
fig.show()
```

```python
# Побудова другої діаграми (наприклад, гістограми)
fig2 = go.Figure(go.Histogram(x=data['median_salary']))

# Налаштування вигляду другої діаграми
fig2.update_layout(title='Середня зарплата по країнах',
                   xaxis_title='Середня зарплатня',
                   yaxis_title='Кількість країн')

# Ініціалізація режиму відображення у зошиті
pio.renderers.default = "colab"

# Створення сітки для двох графіків
figAll = make_subplots(rows=1, cols=2)

# Додавання першої діаграми на перший графік у сітці
figAll.add_trace(fig.data[0], row=1, col=1)

# Додавання другої діаграми на другий графік у сітці
figAll.add_trace(fig2.data[0], row=1, col=2)

# Оновлення заголовків і легенд
figAll.update_layout(title_text='Дві діаграми', showlegend=False)

# Відобразіть сітку графіків
figAll.show()
```
