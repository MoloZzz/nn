```python
import folium
import gradio as gr
from IPython.display import HTML

# Функція для створення мапи та додавання маркерів
def create_airbnb_map(city, coordinates):
    # Створення мапи
    airbnb_map = folium.Map(location=coordinates, zoom_start=12)

    # Додавання маркера
    folium.Marker(location=coordinates, popup=city).add_to(airbnb_map)

    # Збереження мапи як HTML
    html_map_path = "/content/airbnb_map.html"
    airbnb_map.save(html_map_path)

    return html_map_path

# Створення інтерфейсу Gradio
iface = gr.Interface(fn=create_airbnb_map,
                     inputs=["text", "text"],
                     outputs="html",
                     live=True,
                     theme="compact",
                     examples=[["New York City", [40.7128, -74.0060]]])

# Запуск інтерфейсу Gradio
iface.launch()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
train_images, test_images = train_images / 255.0, test_images / 255.0

train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)

model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.fit(train_images, train_labels, epochs=5, validation_data=(test_images, test_labels))

test_loss, test_acc = model.evaluate(test_images, test_labels)
print(f"Точність на тестових даних: {test_acc}")

model.save("fashion_mnist_model.h5")
```

```python
import gradio as gr
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.utils import to_categorical

(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
train_images, test_images = train_images / 255.0, test_images / 255.0
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)

model = tf.keras.models.load_model("fashion_mnist_model.h5")

def classify_image(image):
    image = image / 255.0
    image = image.reshape((1, 28, 28))

    prediction = model.predict(image)


    predicted_class = tf.argmax(prediction[0]).numpy()


    class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
                   "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
    result = {"class_name": class_names[predicted_class], "probability": round(prediction[0][predicted_class] * 100, 2)}
    return result


iface = gr.Interface( fn=classify_image,
                     inputs="image",
                     outputs=["text", "text"],
                     theme="compact",
                     live=True)

iface.launch()
```

```python
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist

%matplotlib inline
(train_images, train_labels), (_, _) = fashion_mnist.load_data()

class_names = ['Футболка/Топ', 'Штани', 'Світшот', 'Плаття', 'Пальто', 'Сандалі', 'Сорочка', 'Кросівки', 'Сумка', 'Чоботи']

plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i]])
plt.show()
```
